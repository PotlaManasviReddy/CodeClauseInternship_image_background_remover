```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
cancer = pd.read_csv('https://github.com/YBIFoundation/Dataset/raw/main/Cancer.csv')
```

```python
cancer.head()
```

```python
cancer.info()
```

```python
cancer.describe()
```

```python
cancer.columns
```

```python
y=cancer['diagnosis']
x=cancer[['radius_mean', 'texture_mean', 'perimeter_mean',
       'area_mean', 'smoothness_mean', 'compactness_mean', 'concavity_mean',
       'concave points_mean', 'symmetry_mean', 'fractal_dimension_mean',
       'radius_se', 'texture_se', 'perimeter_se', 'area_se', 'smoothness_se',
       'compactness_se', 'concavity_se', 'concave points_se', 'symmetry_se',
       'fractal_dimension_se', 'radius_worst', 'texture_worst',
       'perimeter_worst', 'area_worst', 'smoothness_worst',
       'compactness_worst', 'concavity_worst', 'concave points_worst',
       'symmetry_worst', 'fractal_dimension_worst']]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=2529)
```

```python
y_train.shape
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression(max_iter=5000)
```

```python
model.fit(x_train,y_train)
```

```python

# Number of features in x_train
num_features = x_train.shape[1]

# Set up subplots (adjust the number of rows and columns based on the number of features)
fig, axes = plt.subplots(nrows=(num_features // 3) + 1, ncols=3, figsize=(15, 5 * (num_features // 3 + 1)))

# Flatten the axes array for easier iteration
axes = axes.flatten()

# Loop through each feature and plot against y_train
for i, feature in enumerate(x_train.columns):
    axes[i].scatter(x_train[feature], y_train, alpha=0.5)
    axes[i].set_xlabel(feature)
    axes[i].set_ylabel('Diagnosis (0 = Benign, 1 = Malignant)')
    axes[i].set_title(f'{feature} vs Diagnosis')

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15, 10))

for i, feature in enumerate(x_train.columns):
    plt.subplot(5, 6, i+1)  # Adjust subplot grid based on number of features
    sns.histplot(data=x_train, x=feature, hue=y_train, kde=True, palette="Set2", multiple="stack")
    plt.xlabel(feature)
    plt.ylabel('Count')
    plt.title(f'{feature} Histogram')

plt.tight_layout()
plt.show()
```

```python
y_pred=model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
